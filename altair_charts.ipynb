# Altair Charts

```python
import pandas as pd
import altair as alt
```

```python
data = pd.read_csv("derived_data/journals.csv")
data.head()
```

```python
# over 5000 rows so need to run this
alt.data_transformers.disable_max_rows()
```

## Date

```python
data["year"] = data.apply(lambda x: x.date[-4:], axis=1)
data.head()
```

```python
months_list = []
length = data["date"].count()
for i in range(length):
    string = data["date"][i]
    split_str = string.split(" ")
    # print(split_str)
    # print(split_str[0])
    months_list.append(split_str[0])
```

```python
data["month"] = months_list
data.head()
```

## People

```python
data["people_edit"] = data["People"]
data.head()
```

```python
data['people_edit'] = data['people_edit'].str.replace("'", "")
data['people_edit'] = data['people_edit'].str.replace("[", "")
data['people_edit'] = data['people_edit'].str.replace("]", "")
data.head()
```

```python
people_list = data["people_edit"].to_list()
new_pp_list = []

for people in people_list:
    split_str = people.split(", ")
    new_pp_list.append(split_str)

new_pp_list
```

```python
final_people_list = []
for i in new_pp_list:
    for people in i:
        final_people_list.append(people)

final_people_list
```

```python
final_people_list.remove("")
final_people_list
```
